```python
# get all the files of the path,读取文件夹下所有文件名字到一个列表
def readname(filePath):
    # 这里一定要使用绝对路径
    name = os.listdir(filePath)
    return name
```

```python
#列表中随机抽取100个不重复的元素
def selectRandom(names):
    return np.random.choice(names, 100, False)
```

```python
# 读取每个类别下的完整样本
filePath = '/Users/hurrican/Desktop/experiment/Spatial/non_zero_vector/0db/'
folder_names = readname(filePath) #这里一定要使用绝对路径
# del folder_names[0] 
print(folder_names)
```

```python
# 进行随机数据段置0
# 选择1～10的一个数N
# 从0～9随机选择n个数，记为数组nums[N]
# 对数组循环，置零0*nums[N]
def set_zero(arr, save_path,n):
    N = n
    pos = random.sample(range(0,10),N) # generate distinct N numbers
    print(pos)
    for num in pos:
        begin = num * 1024
        end = (num+1)*1024
        # print(begin)
        # print(end)
        for i in range(begin,end):
            arr[i] = [0,0];
    np.save(file = save_path,arr = arr)    
```
